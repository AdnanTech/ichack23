```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import medfilt

# Seed the random number generator for reproducibility
np.random.seed(0)

# Generate time points
t = np.linspace(0, 10, num=1000)

# Generate heart rate data
mean_hr = 60
std_hr = 5
heart_rate = mean_hr + std_hr * np.random.randn(len(t))

# Apply a median filter to smooth the data
smoothed_hr = medfilt(heart_rate, kernel_size=51)

# Plot the original and smoothed heart rate data
plt.plot(t, heart_rate, label='Original')
plt.plot(t, smoothed_hr, label='Smoothed')
plt.xlabel('Time (s)')
plt.ylabel('Heart Rate (bpm)')
plt.legend()
plt.show()
```
